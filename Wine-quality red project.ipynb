**Check for Missing Values**

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(), cbar=False, cmap='Blues')
plt.title("Missing Values in Dataset")
plt.show()
```

**Distribution of Each Feature**

```python
df.hist(bins=15, figsize=(15,10), color='teal', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap", fontsize=14)
plt.show()
```

**Alcohol vs (Target Column)**

```python
import warnings
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='alcohol', data=df, palette='coolwarm')
plt.title("Alcohol Content vs Wine Quality")
plt.show()
```

```python
df.hist(bins=15, figsize=(15,10), color='teal', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

**Countplot of Quality**

```python
sns.countplot(x='quality', data=df, palette='viridis')
plt.title('Count of Each Wine Quality')
plt.show()
```

**Boxplots for Features vs Quality**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Alcohol vs Wine Quality')
plt.show()
```

**Scatter Plot with Regression Line (Alcohol vs Quality)**

```python
plt.figure(figsize=(8,6))
sns.regplot(x='alcohol', y='quality', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Alcohol vs Wine Quality (with Regression Line)', fontsize=14)
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

**Select features most correlated with quality**

```python
top_features = df.corr()['quality'].abs().sort_values(ascending=False)[1:6].index
plt.figure(figsize=(8,6))
sns.heatmap(df[top_features].corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation among Top Features Affecting Quality', fontsize=14)
plt.show()
```
