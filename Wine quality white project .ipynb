**Data Visualisation**

```python
# Set style for plots
sns.set(style="whitegrid")
```

**Distribution of Wine Quality**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=White_wine, palette='viridis')
plt.title("Distribution of White Wine Quality Ratings")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(10,8))
sns.heatmap(White_wine.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of White Wine Features")
plt.show()
```

**Alcohol vs Quality**

```python
plt.figure(figsize=(8,5))
sns.barplot(x='quality', y='alcohol', data=White_wine, palette='magma')
plt.title("Average Alcohol Content by Wine Quality")
plt.show()

```

**Volatile Acidity vs Wine Quality**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='volatile acidity', data=White_wine, palette='Set2')
plt.title("Volatile Acidity vs Wine Quality")
plt.show()
```

**Citric Acid vs Quality**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='citric acid', data=White_wine, palette='cool')
plt.title("Citric Acid vs Wine Quality")
plt.show()
```

**Pairplot for Selected Features**

```python
selected = ['alcohol', 'volatile acidity', 'citric acid', 'density', 'residual sugar', 'quality']
sns.pairplot(White_wine[selected], hue='quality', palette='husl')
plt.show()
```
